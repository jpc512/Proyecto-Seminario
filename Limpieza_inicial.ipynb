```python
#Tratamiento datos primera columna, colocaremos un genero aleatorio para donde no se tenga informacion del genero
import random
df["Genero_Usuario"].astype(str)
mask = df["Genero_Usuario"] == "?"
random_generos = [random.choice(['M', 'F']) for _ in range(mask.sum())]
df.loc[mask, "Genero_Usuario"] = random_generos
df["Genero_Usuario"].value_counts()
```

```python
#ahora vamos a tratar los datos en la segunda columna
df["Edad_Usuario"].unique()
```

```python
#Del resultado anterior vemos que hay valores altos en las edades, que son valores atipicos que consideraremos luego
#Lo que haremos sera rellenar con la media, redondeada al entero mayor mas cercano para rellenar los datos daltantes
df["Edad_Usuario"].isna().sum()
```

```python
media = round(float(df["Edad_Usuario"].mean()),0)
df["Edad_Usuario"] = df["Edad_Usuario"].fillna(media)
df["Edad_Usuario"].isna().sum()
```

```python
df['Hora_Retiro'] = pd.to_timedelta(df['Hora_Retiro'])
df['Hora_Arribo'] = pd.to_timedelta(df['Hora_Arribo'])

# Si la hora de arribo es menor que la hora de retiro, se suma un día (24 horas) al 'Hora_Arribo'
df['Duracion'] = df['Hora_Arribo'] - df['Hora_Retiro']
df.loc[df['Duracion'] < pd.Timedelta(0), 'Duracion'] += pd.Timedelta(days=1)
df['Duracion'] = round((df['Duracion']).dt.total_seconds() / 60,2)
```

```python
#Separar Fecha en año, mes y dia
df["Fecha_Arribo"] = pd.to_datetime(df["Fecha_Arribo"], format="%d/%m/%Y")
df["Año"] = df["Fecha_Arribo"].dt.year
df["Mes"] = df["Fecha_Arribo"].dt.month
df["Dia"] = df["Fecha_Arribo"].dt.day
df
```

```python
# Ensure the date column is in datetime format
df["Fecha_Arribo"] = pd.to_datetime(df["Fecha_Arribo"], format="%d/%m/%Y")

# Extract the day of the week
df["Dia_Semana"] = df["Fecha_Arribo"].dt.day_name()

# Display the updated DataFrame
df.head()
```

```python
df['Dia_Semana'].value_counts()
```

```python
#Obtener hora de la hora de arribo y de retiro, dejar solo la hora
# Convert timedelta to hours by dividing total seconds by 3600 and taking the floor
df['Hora_Arribo'] = (df['Hora_Arribo'].dt.total_seconds() // 3600).astype(int)
```

```python
df['Hora_Retiro'] = (df['Hora_Retiro'].dt.total_seconds() // 3600).astype(int)
df
```

```python
# Drop columnas innecesarias
df.drop(["Fecha_Arribo",'Fecha_Retiro'],axis=1,inplace=True)
```

```python
df['Mes'] = df['Año'].astype(str) + '-' + df['Mes'].astype(str)
df.drop(columns=['Año'], inplace=True)
df.head()
```
