```python
import pandas as pd
meses1 = ['01','02','03','04','05','06','07','08','09','10','11','12']
meses2 = meses1 = ['01','02','03']
archivos_2024 = [f"2024-{i}.csv" for i in meses1]
archivos_2025 = [f"2025-{i}.csv" for i in meses2]
dfs = [pd.read_csv(archivo) for archivo in archivos_2024 + archivos_2025]
df = pd.concat(dfs, ignore_index=True)
df
```

# EDA Estadisticas Comunes

```python
df.info()
```

```python
df["Genero_Usuario"].value_counts()
```

```python
#Tratamiento datos primera columna, colocaremos un genero aleatorio para donde no se tenga informacion del genero
import random
df["Genero_Usuario"].astype(str)
mask = df["Genero_Usuario"] == "?"
random_generos = [random.choice(['M', 'F', 'O']) for _ in range(mask.sum())]
df.loc[mask, "Genero_Usuario"] = random_generos
df["Genero_Usuario"].value_counts()
```

```python
#ahora vamos a tratar los datos en la segunda columna
df["Edad_Usuario"].unique()
```

```python
#Del resultado anterior vemos que hay valores altos en las edades, que son valores atipicos que consideraremos luego
#Lo que haremos sera rellenar con la media, redondeada al entero mayor mas cercano para rellenar los datos daltantes
df["Edad_Usuario"].isna().sum()
```

```python
media = round(float(df["Edad_Usuario"].mean()),0)
df["Edad_Usuario"] = df["Edad_Usuario"].fillna(media)
df["Edad_Usuario"].isna().sum()
```

```python
df["Edad_Usuario"].unique()
```

```python
df["Edad_Usuario"] = pd.to_numeric(df["Edad_Usuario"], errors='coerce').astype("Int64")
df.info()
```

```python
df
```

```python
Fecha_Arribo = []
fecha_arribo1 = df["Fecha Arribo"].tolist()
fecha_arribo2 = df["Fecha_Arribo"].tolist()
longitud = df.shape[0]

for i in range(longitud):
    fecha = ""
    if pd.isna(fecha_arribo1[i]):
        fecha = fecha_arribo2[i]
    else:
        fecha = fecha_arribo1[i]

    Fecha_Arribo.append(fecha)

pd.Series(Fecha_Arribo).isna().sum()
```

```python
df["Fecha_Arribo"] = Fecha_Arribo
df.drop(["Fecha Arribo"],axis=1,inplace=True)
df
```

```python
df["FechaHora_Retiro"] = pd.to_datetime(
    df["Fecha_Retiro"].astype(str) + " " + df["Hora_Retiro"].astype(str),
    dayfirst=True,
    errors="coerce"
)

df["FechaHora_Arribo"] = pd.to_datetime(
    df["Fecha_Arribo"].astype(str) + " " + df["Hora_Arribo"].astype(str),
    dayfirst=True,
    errors="coerce"
)
df
```
