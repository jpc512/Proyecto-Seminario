```python
import pandas as pd
df1 = pd.read_csv("2024-10.csv")
df2 = pd.read_csv("2024-11.csv")
df3 = pd.read_csv("2024-12.csv")
df4 = pd.read_csv("2025-01.csv")
df5 = pd.read_csv("2025-02.csv")
df6 = pd.read_csv("2025-03.csv")
```

```python
dfs = [df1,df2,df3,df4,df5,df6]
column_sets = [set(df.columns) for df in dfs]
all(cols == column_sets[0] for cols in column_sets)
```

```python
df = pd.concat(dfs, ignore_index=True)
df
```

```python
df.columns
```

```python
df['FechaHora_Retiro'] = pd.to_datetime(df['Fecha_Retiro'] + ' ' + df['Hora_Retiro'], format="%d/%m/%Y %H:%M:%S")
df['FechaHora_Arribo'] = pd.to_datetime(df['Fecha_Arribo'] + ' ' + df['Hora_Arribo'], format="%d/%m/%Y %H:%M:%S")

# 2. Calcular duraciÃ³n como timedelta
df['Duracion_Trayecto'] = df['FechaHora_Arribo'] - df['FechaHora_Retiro']

# 3. Crear nuevo DataFrame con las columnas que necesitas
df_nuevo = df[[
    'Genero_Usuario',
    'Edad_Usuario',
    'Bici',
    'Ciclo_Estacion_Retiro',
    'Ciclo_EstacionArribo',
    'Duracion_Trayecto'
]].copy()  # <- esta es la clave
```

```python
# Calcular duraciÃ³n en segundos como nÃºmero
df_nuevo['Duracion_Segundos'] = df['Duracion_Trayecto'].dt.total_seconds()
df_nuevo['Duracion_Minutos'] = df_nuevo['Duracion_Segundos'] / 60
df_nuevo['Duracion_Minutos'] = df_nuevo['Duracion_Minutos'].round(2)
df_nuevo
```

```python
df_nuevo = df_nuevo.drop(["Duracion_Trayecto","Duracion_Segundos"],axis=1)
df_nuevo
```

```python
df_nuevo.rename(columns={
    'Genero_Usuario': 'Genero',
    'Edad_Usuario': 'Edad',
    'Ciclo_Estacion_Retiro': 'Estacion salida',
    'Ciclo_EstacionArribo': 'Estacion llegada',
    'Duracion_Minutos': 'Duracion viaje en minutos'
}, inplace=True)
df_nuevo
```

```python
df_nuevo = df_nuevo.dropna()
df_nuevo
```

```python
df_nuevo = df_nuevo[df_nuevo["Genero"] != "?"]
df_nuevo
```

```python
df_muestra = df_nuevo.sample(frac=0.001, random_state=42)  # 1% de la data (~109,000 filas)
df_muestra
```

```python
# 1. Crear copia del DataFrame original
df_limpio = df_muestra.copy()

# 2. Obtener nombres de columnas numÃ©ricas
columnas_numericas = df_limpio.select_dtypes(include='number').columns

# 3. Aplicar filtro IQR solo a columnas numÃ©ricas
for col in columnas_numericas:
    Q1 = df_limpio[col].quantile(0.25)
    Q3 = df_limpio[col].quantile(0.75)
    IQR = Q3 - Q1
    lim_inf = Q1 - 1.5 * IQR
    lim_sup = Q3 + 1.5 * IQR
    
    # Filtrar solo filas dentro de los lÃ­mites
    df_limpio = df_limpio[(df_limpio[col] >= lim_inf) & (df_limpio[col] <= lim_sup)]

# 4. Resultado final: mismo DataFrame pero sin outliers numÃ©ricos
print("TamaÃ±o original:", df_muestra.shape)
print("TamaÃ±o sin outliers:", df_limpio.shape)
```

```python
df_limpio.to_csv("datos limpios.csv", index=False)
```
